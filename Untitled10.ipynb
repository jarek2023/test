```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 60)
```

```python
columns = ["EmpID","AgeGroup","Attrition","BusinessTravel","Department","Education","EducationField","Gender","HourlyRate","JobRole","MonthlyIncome","PercentSalaryHike","YearsAtCompany","YearsSinceLastPromotion","PerformanceRating","JobSatisfaction"]
hr = pd.read_csv("HR_Analytics.csv", usecols=columns )
```

```python
hr.head(15)
```

```python
hr.info(memory_usage='deep')
```

```python
data_type_dictionary = ({'AgeGroup' : 'category', 'Education':'Int8', 'Department' : 'category', 'Gender' : 'category',
           'JobSatisfaction' : 'Int8', 'MonthlyIncome' : 'float64','PercentSalaryHike' : 'Int16', 
           'PerformanceRating' : 'Int16', 'YearsAtCompany' : 'Int8', 'YearsSinceLastPromotion' : 'Int8' })

hr= hr.astype(data_type_dictionary)
```

```python
hr.info(memory_usage='deep')
```

```python
hr.isna().sum()
```

```python
hr.duplicated().sum()
```

```python
hr[hr.duplicated(['EmpID'], keep=False)]
```

```python
hr = hr.drop_duplicates()
hr.duplicated().sum()
```

```python
hr['BusinessTravel'].unique()
```

```python
hr['BusinessTravel'] = hr['BusinessTravel'].str.replace('_', '')
```

```python
hr.describe()
```

```python
hr.groupby('Department', observed=True).agg({'MonthlyIncome' : ['mean','min','max']})
```
